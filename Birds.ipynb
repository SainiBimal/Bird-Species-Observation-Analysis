**Sheets details:**

The Excel file contains multiple sheets representing different administrative units, with their codes matching the Admin_Unit_Code column:

---

**Columns details:**

It includes detailed columns describing location, observation methods, bird species, and environmental conditions.


Below are the key columns present in each sheet of the Excel files

- **Admin_Unit_Code:** Code for the administrative unit (e.g., ANTI, CATO).

- **Sub_Unit_Code:** Subdivision within the administrative unit.
- **Site_Name:** Name of the specific observation site.
- **Plot_Name:** Unique identifier for the observation plot.
- **Location_Type:** Type of habitat (e.g., Forest, Grassland).
- **Year:** Year when the observation was made.
- **Date:** Exact date of observation (in YYYY-MM-DD format).
- **Start_Time:** Time when the observation session started.
- **End_Time:** Time when the observation session ended.
- **Observer:** Person who recorded the observation.
- **Visit:** Number of times the observer visited the site.
- **Interval_Length:** Duration of the observation interval (e.g., 0–2.5 min).
- **ID_Method:** Method used to identify species (e.g., Singing, Calling, Visualization).
- **Distance:** Distance of the observed bird from the observer (e.g., ≤ 50 meters).
- **Flyover_Observed:** Indicates if the bird was seen flying overhead (TRUE/FALSE).
- **Sex:** Gender of the observed bird (Male, Female, Undetermined).
- **Common_Name:** Common name of the bird species.
- **Scientific_Name:** Scientific (Latin) name of the bird species.
- **AcceptedTSN:** Taxonomic Serial Number for the species.
- **NPSTaxonCode:** Unique code assigned to the species in the NPS system.
- **AOU_Code:** American Ornithological Union code for the species.
- **PIF_Watchlist_Status:** Indicates if the species is on the Partners in Flight Watchlist (TRUE/FALSE).
- **Regional_Stewardship_Status:** Whether the species has regional conservation priority (TRUE/FALSE).
- **Temperature:** Recorded temperature during the observation.
- **Humidity:** Recorded humidity during the observation.
- **Sky:** Sky condition at the time (e.g., Clear, Overcast).
- **Wind:** Wind condition during the observation.
- **Disturbance:** Any disturbance that may have impacted the observation (e.g., No effect on count).
- **Initial_Three_Min_Cnt:** Number of birds observed in the first 3 minutes of the session.

# **Data Wrangling**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
from pandas.errors import SettingWithCopyWarning

warnings.filterwarnings("ignore", category=SettingWithCopyWarning)
```

```python
# Step 1: Load all sheets from both files
forest_sheets = pd.read_excel("/content/drive/MyDrive/Bird Species Observation Analysis/Bird_Monitoring_Data_FOREST.XLSX", sheet_name=None)
grassland_sheets = pd.read_excel("/content/drive/MyDrive/Bird Species Observation Analysis/Bird_Monitoring_Data_GRASSLAND.XLSX", sheet_name=None)
```

```python
print(forest_sheets)
print(grassland_sheets)
```

```python
# Step 2: Combine all forest sheets
forest = pd.concat(forest_sheets.values(), ignore_index=True)

# Step 3: Combine all grassland sheets
grassland= pd.concat(grassland_sheets.values(), ignore_index=True)


forest
grassland
```

```python
# Step 4: Combine both into one final DataFrame
bird = pd.concat([forest, grassland], ignore_index=True)
bird
```

```python
# Imputing Null Values
bird["Sub_Unit_Code"] = bird["Sub_Unit_Code"].fillna("Unknown")
bird["Site_Name"] = bird["Site_Name"].fillna("Unknown")
bird["ID_Method"] = bird["ID_Method"].fillna(bird["ID_Method"].mode()[0])
bird["Distance"] = bird["Distance"].fillna(bird["Distance"].mode()[0])
bird["Sex"] = bird["Sex"].fillna("Unknown")
bird["AcceptedTSN"] = bird["AcceptedTSN"].fillna("Unknown")
bird["NPSTaxonCode"] = bird["NPSTaxonCode"].fillna("Unknown")
bird["TaxonCode"] = bird["TaxonCode"].fillna("Unknown")
bird["Previously_Obs"] = bird["Previously_Obs"].fillna("Unknown")
```

```python
bird.isnull().sum()
```

```python
# Change the datatypes

bird["Flyover_Observed"] = pd.to_numeric(bird["Flyover_Observed"], errors="coerce")
bird["PIF_Watchlist_Status"] = pd.to_numeric(bird["PIF_Watchlist_Status"], errors = "coerce")
bird["Regional_Stewardship_Status"] = pd.to_numeric(bird["Regional_Stewardship_Status"], errors = "coerce")
bird["Initial_Three_Min_Cnt"] = pd.to_numeric(bird["Initial_Three_Min_Cnt"], errors="coerce")
bird["Year"] = bird["Year"].astype(int)
bird["Start_Time"] = pd.to_datetime(bird["Start_Time"], format = "%H:%M:%S")
bird["End_Time"] = pd.to_datetime(bird["End_Time"], format = "%H:%M:%S")
bird.dtypes
```

```python
bird.head()
```

```python
# Checking For Duplicates
bird.duplicated().sum()
```

```python
# Drop the Duplicate values
bird = bird.drop_duplicates()
```

```python
# Feature Engineering adding new column to the datasets

bird["Hour"] = pd.to_datetime(bird["Start_Time"]).dt.hour
bird["Minute_Duration"] = ((bird["End_Time"]) - (bird["Start_Time"])).dt.total_seconds() / 60
bird["Month_Num"] = bird["Date"].dt.month


bird["Month"] = bird["Date"].dt.month_name()



bird["Temperature_Interval"] = pd.cut(bird["Temperature"], bins = 10)

bird["Humidity_Interval"] = pd.cut(bird["Humidity"], bins =10)


bird.loc[bird["Month"].isin(["March", "April", "May"]), "Season"] = "Summer"
bird.loc[bird["Month"].isin(["June", "July", "August"]), "Season"] = "Monsoon"
bird.loc[bird["Month"].isin(["September", "October", "November"]), "Season"] = "Autumn"
bird.loc[bird["Month"].isin(["December", "January", "February"]), "Season"] = "Winter"
```
